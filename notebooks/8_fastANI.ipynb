```python
import os
import pandas as pd
import numpy as np
from Bio import SeqIO
from Bio.SeqRecord import SeqRecord
```

## ANI between 12 organisms

### Sort aligned 92 genes to 12 organisms

```python
path = '../data/fastANI_data/pre_aligned_genes_by_orgs/'
gene_names = [f.split('.')[0] for f in os.listdir(path)]
full_paths = [path + f for f in os.listdir(path)]
```

```python
gene_names[:10]
```

```python
full_paths[:5]
```

```python
org_ids = []
orgs = {}

for gene_file, gene_name in zip(full_paths, gene_names):
    genes = SeqIO.parse(gene_file, 'fasta')
    for gene in genes:
        org_ids.append(gene.id)
        if orgs.get(gene.id) is None:
            orgs[gene.id] = []
        orgs[gene.id].append((gene_name, gene.seq))
```

```python
org_ids = list(set(org_ids))
```

```python
path_orgs = '../data/fastANI_data/orgs_gene_aligned_1066/'

for org_id in org_ids:
    write_path = path_orgs + org_id + '.fasta'
    records = []
    
    for (gene_id, gene_seq) in orgs[org_id]:
        records.append(SeqRecord(gene_seq, id=gene_id, description='| '+org_id))
    
    with open(write_path, 'w') as output_file:
        SeqIO.write(records, output_file, 'fasta')
```

```python
full_paths_orgs = [path_orgs + f for f in os.listdir(path_orgs) if f.endswith('.fasta')]
```

## ANI between similar genes (92 genes)

```python
path = '../data/fastANI_data/pre_aligned_genes_by_orgs/'
gene_names = [f.split('.')[0] for f in os.listdir(path)]
full_paths = [path + f for f in os.listdir(path)]
```

```python
gene_names[:10]
```

```python
path_genes_output = '../data/fastANI_data/genes_unique_1066/'

for gene_file, gene_name in zip(full_paths, gene_names):
    genes = SeqIO.parse(gene_file, 'fasta')
    for gene in genes:
        gene_org_id = gene_name+'_'+'_'.join(gene.id.split('_')[:2])
        write_path = path_genes_output + gene_org_id + '.fasta'
        
        with open(write_path, 'w') as output_file:
            SeqIO.write(SeqRecord(gene.seq, id=gene_org_id, description='|'), output_file, 'fasta')
```

```python
full_paths_genes_by_one = [path_genes_output + f for f in os.listdir(path_genes_output) if f.endswith('.fasta')]
```

```python
genes_by_groups = []

for gene_name in gene_names:
    one_gene_group = []
    for f in full_paths_genes_by_one:
        if f.split('/')[2].startswith(gene_name):
            one_gene_group.append(f)
            
    genes_by_groups.append((gene_name, one_gene_group))
```

```python
genes_by_groups[:3]
```

```python
path_fastani_output = '../data/fastANI_data/genes_unique_1066_fastANI/'
```

```python
# one by one
# rather long!

fastani_output_dirs = []

frag_len = 100
kmers = 8
min_frag = 5

for gene_group in genes_by_groups:
    print(gene_group[0])
    output_dir = path_fastani_output + gene_group[0]
    fastani_output_dirs.append(output_dir)
    # first of all - make 92 directories
    !mkdir {output_dir}
    for (p1, path_gene1) in enumerate(gene_group[1]):
        for (p2, path_gene2) in enumerate(gene_group[1]):
            
            output_file = output_dir + '/output_' + str(p1+1) + '_' + str(p2+1) + '.txt'
            
            !fastANI -q {path_gene1} -r {path_gene2} -o {output_file} --fragLen {frag_len} -k {kmers} --minFrag {min_frag}
```

### Show fastANI results between similar genes

```python
len(fastani_output_dirs), fastani_output_dirs[:10]
```

```python
for d in fastani_output_dirs:
    paths_gene_fastani = [d + '/' + f for f in os.listdir(d) if f.endswith('.txt')]
    for file_name in paths_gene_fastani:
        with open(file_name) as file:
            data = file.read()
            if data != '':
                data = data.split("\t")
                gene1 = data[0].split("/")[2].split(".")[0]
                gene2 = data[1].split("/")[2].split(".")[0]
                ani = data[2]
                print(f'{gene1} - {gene2}: {ani}\n')
```
